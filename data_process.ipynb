# color 처리

- color 기준으로 이름도 변경

```python
color_0=glob('./data/2022_05/Color/color_0/*.jpg')
color_1=glob('./data/2022_05/Color/color_1/*.jpg')
color_2=glob('./data/2022_05/Color/color_2/*.jpg')

print("color_0 : ",len(color_0))
print("color_1 : ",len(color_1))
print("color_2 : ",len(color_2))
print("총 샘플 수 :",len(color_0)+len(color_1)+len(color_2) )
```

```python
#이름 생성기

colors_list = [color_0,color_1,color_2]

index_count=0

name_dict = dict()

for colors in colors_list:
    origin_names = [color.split("\\")[1].split('.')[0] for color in colors ]
    change_names = []
    for name in origin_names:
        print(name,"->",str(index_count))
        name_dict[name] = str(index_count)
        index_count+=1
```

```python

name_dict
```

```python
# 파일 이름 변경 color_0

color_dict = dict()

for name in color_0:
    src = name.split("\\")[0]
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.jpg'
    dst = os.path.join(src, dst)
    color_dict[name_dict[ name.split('\\')[1].split(".")[0] ]] = 0
    print(dst)    
    os.rename(name, dst)
    
    #json파일
    name_json = '.'+name.split('.')[1]+'.json' 
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.json'
    dst = os.path.join(src, dst)
    print(dst)
    os.rename(name_json, dst)    
```

```python
# 파일 이름 변경 color_1,
for name in color_1:
    src = name.split("\\")[0]
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.jpg'
    dst = os.path.join(src, dst)
    color_dict[name_dict[ name.split('\\')[1].split(".")[0] ]] = 1
    print(dst)    
    os.rename(name, dst)
    
    #json파일
    name_json = '.'+name.split('.')[1]+'.json' 
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.json'
    dst = os.path.join(src, dst)
    print(dst)
    os.rename(name_json, dst)    
```

```python
# 파일 이름 변경 color_2,
for name in color_2:
    src = name.split("\\")[0]
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.jpg'
    dst = os.path.join(src, dst)
    color_dict[name_dict[ name.split('\\')[1].split(".")[0] ]] = 2
    print(dst)    
    os.rename(name, dst)
    
    #json파일
    name_json = '.'+name.split('.')[1]+'.json' 
    dst = str(name_dict[ name.split('\\')[1].split(".")[0] ]) + '.json'
    dst = os.path.join(src, dst)
    print(dst)
    os.rename(name_json, dst)    
```

```python
#color 따로 excel 만들기


new_name_list =[]
colors_list = []


for origin_name in list(name_dict.keys()):
    new_name = name_dict[origin_name]
    print(new_name)
    new_name_list.append(new_name)
    colors_list.append(color_dict[new_name])

    
color_dataframe= pd.concat([
                            pd.DataFrame(list(name_dict.keys()) ),
                            pd.DataFrame(new_name_list),
                            pd.DataFrame(colors_list)
                        ],axis=1)
color_dataframe.columns=['origin_name','file_index','color']
```

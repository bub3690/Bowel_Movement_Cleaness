```python
#데이터 제거 
for file_name in bmc_duplicated_df.loc[ bmc_duplicated_df['label']==0]['file']:
    file_head = '../data/2022_10_preprocess/0/'
    image_file_name= file_head+file_name+'.jpg'
    json_file_name= file_head+file_name+'.json'
    
    if  os.path.isfile(image_file_name):
        os.remove(image_file_name)
    if  os.path.isfile(json_file_name):
        os.remove(json_file_name)
```

```python
#데이터 제거 
for file_name in bmc_duplicated_df.loc[ bmc_duplicated_df['label']==1]['file']:
    file_head = '../data/2022_10_preprocess/1/'
    image_file_name= file_head+file_name+'.jpg'
    json_file_name= file_head+file_name+'.json'
    
    if  os.path.isfile(image_file_name):
        os.remove(image_file_name)
    else:
        print(image_file_name," 제거 실패")
    if  os.path.isfile(json_file_name):
        os.remove(json_file_name)
    else:
        print(json_file_name," 제거 실패")
```

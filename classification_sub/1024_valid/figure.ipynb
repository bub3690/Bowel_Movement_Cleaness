```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
summary_df=pd.read_excel('./baseline_summary.xlsx',sheet_name='baseline_auc',index_col=0)
summary_df
```

```python
plt.plot(summary_df[['roc-auc']])
plt.title('ROC-AUC')
plt.xlabel('Train Sample Size')

for x,y in zip(summary_df.index.tolist(),summary_df['roc-auc'].tolist()):
    text = "{:.2f}".format(y)
    plt.text(x,y+0.001,text)
```

```python
plt.plot(summary_df[['Acc']])
plt.title('ACC')
plt.xlabel('Train Sample Size')


for x,y in zip(summary_df.index.tolist(),summary_df['Acc'].tolist()):
    text = "{:.2f}%".format(y*100)
    plt.text(x,y+0.001,text)
```

```python
plt.plot(summary_df[['f1 score']])
plt.title('f1-score')
plt.xlabel('Train Sample Size')
for x,y in zip(summary_df.index.tolist(),summary_df['f1 score'].tolist()):
    text = "{:.2f}".format(y)
    plt.text(x,y+0.001,text)
```
